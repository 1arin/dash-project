```python
import pandas as pd
```

```python
data = pd.read_csv('clean_data.csv')
data = data.drop(columns='Unnamed: 0')
data['DATETIMEDATA'] = pd.to_datetime(data['DATETIMEDATA'])
data = data.drop(columns=['PM25','PM10','O3','CO','NO2','SO2'])
data
```

```python
from pycaret.regression import *
setup101 = setup(data, target='WD', session_id = 123, date_features=['DATETIMEDATA'], normalize=True)
best_model = compare_models()
tuned = tune_model(best_model)
bagged = ensemble_model(tuned , n_estimators=15)
```

```python
save_model(bagged, 'WD_catboost_pipeline')
load_model('WD_catboost_pipeline')
predict_model(bagged)
```
